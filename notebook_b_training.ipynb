```python
# =============================================================================
# NOTEBOOK B — TRAINING
#
# Changes vs previous version:
#   [DATA]    Loads mobilenet_feats (N, 176) from subject_CSI*.npz
#   [INPUT]   clip (512) + mobilenet (176) concatenated → 688-d per image
#   [ARCH]    img_proj now Linear(688, D) instead of Linear(512, D)
#   [ARCH]    d_model 256→384, n_attn_layers 3→4, mlp_hidden 1024→2048
#   [ARCH]    mlp_n_layers now a CFG parameter (default 4, was hardcoded 3)
#             _build_predictor constructs a depth-configurable MLP:
#               layer 1       : in_dim → hidden  (LayerNorm + GELU + Dropout)
#               layers 2..n-1 : hidden → hidden  (GELU + Dropout)
#               layer n       : hidden → out_dim (raw regression, no activation)
#   [FIX]     mobilenet_dim corrected 184 → 176 (actual torchvision output)
#   [EVAL]    evaluate() and Yeo-7 breakdown pass mobilenet tensors
#   [MULTI-GPU] DataParallel; model.module unwrap handled throughout
# =============================================================================


# ── CELL 0 : INSTALL ──────────────────────────────────────────────────────────
# Run once, then comment out:
# import subprocess
# subprocess.run(["pip", "install", "-q", "scipy", "matplotlib"], check=True)


# ── CELL 1 : IMPORTS ──────────────────────────────────────────────────────────
import json, math, os, random, time
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
from tqdm.auto import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

print(f"PyTorch {torch.__version__}")
print(f"CUDA devices : {torch.cuda.device_count()}")
for i in range(torch.cuda.device_count()):
    p = torch.cuda.get_device_properties(i)
    print(f"  GPU {i}: {p.name}  {p.total_memory/1e9:.1f} GB")


# ── CELL 2 : CONFIG ───────────────────────────────────────────────────────────
CFG = {
    # ── Paths ──────────────────────────────────────────────────────────────
    "preprocessed_dir":    "/kaggle/input/datasets/shadowdavid/bold-dataset/preprocessed",
    "output_dir":          "/kaggle/working/model",

    # ── Data ───────────────────────────────────────────────────────────────
    "subjects":            ["CSI1", "CSI2", "CSI3", "CSI4"],

    # ── Feature dims — must match Notebook A output exactly ────────────────
    "clip_dim":            512,   # MobileCLIP-S0
    "mobilenet_dim":       176,   # MobileNetV3-Small multi-layer GAP
                                  # (features[1,3,6,10] → 16+24+40+96)
                                  # NOTE: paper table says 48 at features[6]
                                  # but actual torchvision output is 40 there,
                                  # giving 176 not 184. Runtime-confirmed.
    "bold_dim":            1024,  # DiFuMo1024 parcels

    # ── Architecture ───────────────────────────────────────────────────────
    "d_model":             384,   # internal attention dim; 384 ÷ 8 heads = 48 ✓
    "n_attn_heads":        8,
    "n_attn_layers":       4,
    "n_fourier_feats":     16,

    # Predictor MLP
    "mlp_hidden":          2048,  # width of all hidden layers
    "mlp_n_layers":        4,     # total Linear layers in BOLD predictor
                                  #   2 = minimal (in→hidden→out)
                                  #   3 = original hardcoded design
                                  #   4 = recommended: dedicated mix/compress/decode
    "dropout":             0.12,

    # ── Meta-learning / calibration ─────────────────────────────────────────
    "n_calib_options":     [50, 100, 200, 500],
    "n_calib_eval":        30,

    # ── Augmentation ────────────────────────────────────────────────────────
    "n_views_per_image":   50,

    # ── Training ────────────────────────────────────────────────────────────
    "epochs":              25,
    "batch_size":          256,
    "lr":                  3e-4,
    "weight_decay":        1e-4,
    "grad_clip":           1.0,
    "warmup_epochs":       5,
    "use_amp":             True,

    # ── Early stopping ───────────────────────────────────────────────────────
    "early_stop_patience":  15,
    "early_stop_min_delta": 1e-5,

    # ── Reproducibility ──────────────────────────────────────────────────────
    "seed":                42,
    "num_workers":          2,

    # ── Device ───────────────────────────────────────────────────────────────
    "device":              "cuda" if torch.cuda.is_available() else "cpu",
}

# Derived constant — never hardcode 688 anywhere else
CFG["img_feat_dim"] = CFG["clip_dim"] + CFG["mobilenet_dim"]   # 688

os.makedirs(CFG["output_dir"], exist_ok=True)

random.seed(CFG["seed"])
np.random.seed(CFG["seed"])
torch.manual_seed(CFG["seed"])
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(CFG["seed"])

print(f"\nPreprocessed dir : {CFG['preprocessed_dir']}")
print(f"Output dir       : {CFG['output_dir']}")
print(f"Device           : {CFG['device']}")
print(f"img_feat_dim     : {CFG['img_feat_dim']}"
      f"  (clip {CFG['clip_dim']} + mobilenet {CFG['mobilenet_dim']})")
print(f"mlp_n_layers     : {CFG['mlp_n_layers']}")


# ── CELL 3 : DATA LOADING ─────────────────────────────────────────────────────

def load_all_subjects(preprocessed_dir: str, subjects: list) -> tuple:
    """
    Returns
    -------
    subject_data : dict  { subj_id → {clips, mobilenets, bolds, scores, is_test} }
    test_names   : set[str]
    yeo_assign   : (1024,) int8
    """
    pdir = Path(preprocessed_dir)

    with open(pdir / "test_image_names.json") as f:
        test_names = set(json.load(f))
    print(f"  Shared held-out test images: {len(test_names)}")

    subject_data = {}
    yeo_assign   = None

    for subj in subjects:
        fpath = pdir / f"subject_{subj}.npz"
        if not fpath.exists():
            print(f"  WARNING: {fpath} not found — skipping {subj}")
            continue

        d = np.load(fpath, allow_pickle=True)

        if yeo_assign is None:
            yeo_assign = d["yeo_assignment"]

        # Hard validation — catches dim mismatches before training starts
        if "mobilenet_feats" not in d:
            raise KeyError(
                f"{subj}: 'mobilenet_feats' missing from {fpath}. "
                "Re-run Notebook A to regenerate preprocessed files."
            )
        mn_shape = d["mobilenet_feats"].shape
        if mn_shape[1] != CFG["mobilenet_dim"]:
            raise ValueError(
                f"{subj}: mobilenet_feats dim={mn_shape[1]}, "
                f"expected {CFG['mobilenet_dim']}. "
                "Update CFG['mobilenet_dim'] to match."
            )

        n_total = len(d["clip_embeddings"])
        n_train = int((~d["is_test"]).sum())
        n_test  = int(d["is_test"].sum())
        print(f"  {subj}: total={n_total}  train={n_train}  test={n_test}"
              f"  clips={d['clip_embeddings'].shape}"
              f"  mobilenets={d['mobilenet_feats'].shape}"
              f"  bolds={d['bold_targets'].shape}")

        subject_data[subj] = {
            "clips":      d["clip_embeddings"].astype(np.float32),   # (N, 512)
            "mobilenets": d["mobilenet_feats"].astype(np.float32),   # (N, 176)
            "bolds":      d["bold_targets"].astype(np.float32),      # (N, 1024)
            "scores":     d["limbic_score"].astype(np.float32),      # (N,)
            "img_names":  np.array(d["img_names"]),
            "is_test":    d["is_test"].astype(bool),
        }

    return subject_data, test_names, yeo_assign


print("\nLoading preprocessed data …")
subject_data, TEST_NAMES, YEO_ASSIGN = load_all_subjects(
    CFG["preprocessed_dir"], CFG["subjects"]
)


# ── CELL 4 : DATASET ──────────────────────────────────────────────────────────

class MultiViewMetaDataset(Dataset):
    """
    Meta-learning dataset with multi-view calibration augmentation.

    Each item returns:
        calib_clips       (N_calib, 512)
        calib_mobilenets  (N_calib, 176)
        calib_scores      (N_calib,)
        target_clip       (512,)
        target_mobilenet  (176,)
        target_bold       (1024,)

    Each target image appears n_views times per epoch, each time paired
    with a freshly sampled calibration set.  The BOLD target never changes;
    only the context does.  This forces the model to learn that calibration
    encodes WHO, not what the target image is.
    """

    def __init__(self, subject_data: dict, n_calib_options: list,
                 n_views: int = 2, split: str = "train"):
        self.n_calib_options = n_calib_options
        self.n_views         = n_views
        self.split           = split

        self.subj_data = {}
        for subj, data in subject_data.items():
            mask    = ~data["is_test"] if split == "train" else data["is_test"]
            indices = np.where(mask)[0]
            if len(indices) == 0:
                continue
            self.subj_data[subj] = {
                "indices":    indices,
                "clips":      data["clips"],
                "mobilenets": data["mobilenets"],
                "bolds":      data["bolds"],
                "scores":     data["scores"],
            }

        self.flat_index = []
        for subj, sd in self.subj_data.items():
            for pos in range(len(sd["indices"])):
                for view in range(n_views):
                    self.flat_index.append((subj, pos, view))

        rng = random.Random(42)
        rng.shuffle(self.flat_index)

        n_subjects = len(self.subj_data)
        n_images   = sum(len(sd["indices"]) for sd in self.subj_data.values())
        print(f"  MultiViewMetaDataset [{split}]:")
        print(f"    Subjects      : {n_subjects}")
        print(f"    Unique images : {n_images}")
        print(f"    Views/image   : {n_views}")
        print(f"    Items/epoch   : {len(self.flat_index)}")

    def __len__(self):
        return len(self.flat_index)

    def __getitem__(self, idx: int) -> dict:
        subj, target_pos, _view = self.flat_index[idx]
        sd      = self.subj_data[subj]
        indices = sd["indices"]
        n       = len(indices)

        n_calib = random.choice(self.n_calib_options)
        n_calib = min(n_calib, n - 1)

        pool            = list(range(n))
        pool.remove(target_pos)
        calib_positions = random.sample(pool, n_calib)

        calib_rows = indices[calib_positions]
        target_row = indices[target_pos]

        return {
            "n_calib":           n_calib,
            "calib_clips":       sd["clips"][calib_rows].copy(),
            "calib_mobilenets":  sd["mobilenets"][calib_rows].copy(),
            "calib_scores":      sd["scores"][calib_rows].copy(),
            "target_clip":       sd["clips"][target_row].copy(),
            "target_mobilenet":  sd["mobilenets"][target_row].copy(),
            "target_bold":       sd["bolds"][target_row].copy(),
        }


def meta_collate(batch: list) -> dict:
    """Pad variable-length calibration sequences; True in pad_mask = ignored."""
    max_n    = max(item["n_calib"] for item in batch)
    B        = len(batch)
    clip_dim = batch[0]["calib_clips"].shape[1]
    mn_dim   = batch[0]["calib_mobilenets"].shape[1]

    calib_clips      = torch.zeros(B, max_n, clip_dim, dtype=torch.float32)
    calib_mobilenets = torch.zeros(B, max_n, mn_dim,   dtype=torch.float32)
    calib_scores     = torch.zeros(B, max_n, 1,        dtype=torch.float32)
    pad_mask         = torch.ones(B,  max_n,           dtype=torch.bool)

    target_clips      = []
    target_mobilenets = []
    target_bolds      = []

    for i, item in enumerate(batch):
        n = item["n_calib"]
        calib_clips[i,      :n] = torch.from_numpy(item["calib_clips"])
        calib_mobilenets[i, :n] = torch.from_numpy(item["calib_mobilenets"])
        calib_scores[i,     :n] = torch.from_numpy(item["calib_scores"]).unsqueeze(-1)
        pad_mask[i,         :n] = False

        target_clips.append(torch.from_numpy(item["target_clip"]))
        target_mobilenets.append(torch.from_numpy(item["target_mobilenet"]))
        target_bolds.append(torch.from_numpy(item["target_bold"]))

    return {
        "calib_clips":       calib_clips,                        # (B, max_n, 512)
        "calib_mobilenets":  calib_mobilenets,                   # (B, max_n, 176)
        "calib_scores":      calib_scores,                       # (B, max_n, 1)
        "pad_mask":          pad_mask,                           # (B, max_n) bool
        "target_clips":      torch.stack(target_clips),          # (B, 512)
        "target_mobilenets": torch.stack(target_mobilenets),     # (B, 176)
        "target_bolds":      torch.stack(target_bolds),          # (B, 1024)
    }


# ── CELL 5 : MODEL ────────────────────────────────────────────────────────────
#
# Architecture:
#
#  rank_score ──► FourierRankEmbed ──► (B, N, D)  ╗
#  calib_clips  ──► cat → (B,N,688) ──► img_proj ──╠─ ADD ──► calib_tokens (B,N,D)
#  calib_mobilenets ──────────────────────────────╝
#
#  new_clip     ──► cat → (B, 688) ──► img_proj ──► query_feat (B,D) ──► unsqueeze ──► (B,1,D)
#  new_mobilenet ──┘                                     │
#                                                        │
#                                               cross-attn × L layers
#                                                        │
#                                                  context (B, D)
#                                                        │
#                                        [query_feat || context] (B, 2D)
#                                                        │
#                                              _build_predictor MLP
#                                                        │
#                                              BOLD pred (B, 1024)

class FourierRankEmbed(nn.Module):
    """
    Encodes a scalar rank score ∈ [0,1] via Fourier features → d_model.
    Fixed frequencies; the linear layers do the learning.
    """
    def __init__(self, n_fourier: int, d_model: int):
        super().__init__()
        freqs = torch.arange(1, n_fourier + 1, dtype=torch.float32) * math.pi
        self.register_buffer("freqs", freqs)
        self.proj = nn.Sequential(
            nn.Linear(2 * n_fourier, d_model),
            nn.GELU(),
            nn.Linear(d_model, d_model),
        )

    def forward(self, score: torch.Tensor) -> torch.Tensor:
        """score : (B, N, 1) → (B, N, d_model)"""
        s     = score * self.freqs
        feats = torch.cat([s.sin(), s.cos()], dim=-1)
        return self.proj(feats)


class CalibratedBoldPredictor(nn.Module):
    """
    Predicts whole-brain BOLD response to a new image, personalised by
    (image, aesthetic-rank) calibration pairs.

    Input per image : clip (512) + mobilenet (176) = 688-d.

    CFG keys consumed:
        img_feat_dim   : 688
        bold_dim       : 1024
        d_model        : 384
        n_attn_heads   : 8
        n_attn_layers  : 4
        n_fourier_feats: 16
        mlp_hidden     : 2048
        mlp_n_layers   : 4
        dropout        : 0.12
    """

    def __init__(self, cfg: dict):
        super().__init__()
        D  = cfg["d_model"]
        I  = cfg["img_feat_dim"]      # 688
        B  = cfg["bold_dim"]
        H  = cfg["n_attn_heads"]
        L  = cfg["n_attn_layers"]
        F_ = cfg["n_fourier_feats"]
        MH = cfg["mlp_hidden"]
        NL = cfg["mlp_n_layers"]
        P  = cfg["dropout"]

        assert D % H == 0, f"d_model={D} must be divisible by n_heads={H}"

        # ── Shared image projector (clip + mobilenet → d_model) ───────────────
        self.img_proj = nn.Sequential(
            nn.Linear(I, D),
            nn.LayerNorm(D),
        )

        # ── Rank encoder ─────────────────────────────────────────────────────
        self.rank_embed = FourierRankEmbed(n_fourier=F_, d_model=D)

        # ── Prior token (fallback when calibration is sparse) ─────────────────
        self.prior_token = nn.Parameter(torch.randn(1, 1, D) * 0.02)

        # ── Cross-attention personalisation stack ─────────────────────────────
        self.cross_layers = nn.ModuleList()
        for _ in range(L):
            self.cross_layers.append(nn.ModuleDict({
                "attn": nn.MultiheadAttention(
                    embed_dim=D, num_heads=H, dropout=P, batch_first=True
                ),
                "ln_q":   nn.LayerNorm(D),
                "ln_kv":  nn.LayerNorm(D),
                "ffn": nn.Sequential(
                    nn.Linear(D, D * 4), nn.GELU(), nn.Dropout(P),
                    nn.Linear(D * 4, D),
                ),
                "ln_ffn": nn.LayerNorm(D),
            }))

        # ── BOLD predictor MLP ────────────────────────────────────────────────
        self.predictor = self._build_predictor(
            in_dim   = D * 2,   # [query_feat (D) || context (D)]
            hidden   = MH,
            out_dim  = B,
            n_layers = NL,
            dropout  = P,
        )

        self._init_weights()

    # ── Predictor builder ─────────────────────────────────────────────────────
    @staticmethod
    def _build_predictor(in_dim: int, hidden: int, out_dim: int,
                         n_layers: int, dropout: float) -> nn.Sequential:
        """
        Build the BOLD predictor MLP with configurable depth.

        Structure:
          Layer 1       : in_dim → hidden  (LayerNorm + GELU + Dropout)
          Layers 2..n-1 : hidden → hidden  (GELU + Dropout)
          Layer n       : hidden → out_dim (no activation — raw regression)

        n_layers=2 : in → hidden → out             (minimal)
        n_layers=3 : in → hidden → hidden → out    (original hardcoded design)
        n_layers=4 : in → h → h → h → out          (recommended)
        """
        if n_layers < 2:
            raise ValueError(f"mlp_n_layers must be >= 2, got {n_layers}")

        layers = []

        # First layer — project up from [query_feat || context]
        layers += [
            nn.Linear(in_dim, hidden),
            nn.LayerNorm(hidden),
            nn.GELU(),
            nn.Dropout(dropout),
        ]

        # Middle layers — constant-width hidden processing
        for _ in range(n_layers - 2):
            layers += [
                nn.Linear(hidden, hidden),
                nn.GELU(),
                nn.Dropout(dropout),
            ]

        # Final layer — project to BOLD parcel space
        layers.append(nn.Linear(hidden, out_dim))

        return nn.Sequential(*layers)

    # ── Weight init ───────────────────────────────────────────────────────────
    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
        nn.init.normal_(self.prior_token, std=0.02)

    # ── Helpers ───────────────────────────────────────────────────────────────
    def _project_image(self,
                       clips:      torch.Tensor,   # (..., 512)
                       mobilenets: torch.Tensor    # (..., 176)
                      ) -> torch.Tensor:            # (..., D)
        """Concat clip + mobilenet then project. Works for 2-D and 3-D inputs."""
        return self.img_proj(torch.cat([clips, mobilenets], dim=-1))

    def _build_kv(self,
                  calib_clips:      torch.Tensor,   # (B, N, 512)
                  calib_mobilenets: torch.Tensor,   # (B, N, 176)
                  calib_scores:     torch.Tensor    # (B, N, 1)
                 ) -> torch.Tensor:                  # (B, N+1, D)
        """
        Build Key/Value tokens: [prior_token | calibration_tokens].
        Each calibration token = projected image features + additive rank embed.
        """
        img_feats  = self._project_image(calib_clips, calib_mobilenets)  # (B, N, D)
        rank_feats = self.rank_embed(calib_scores)                        # (B, N, D)
        calib_tok  = img_feats + rank_feats                               # (B, N, D)

        B     = calib_clips.shape[0]
        prior = self.prior_token.expand(B, -1, -1)                        # (B, 1, D)
        return torch.cat([prior, calib_tok], dim=1)                       # (B, N+1, D)

    # ── Forward ───────────────────────────────────────────────────────────────
    def forward(self,
                new_clip:         torch.Tensor,   # (B, 512)
                new_mobilenet:    torch.Tensor,   # (B, 176)
                calib_clips:      torch.Tensor,   # (B, N, 512)
                calib_mobilenets: torch.Tensor,   # (B, N, 176)
                calib_scores:     torch.Tensor,   # (B, N, 1)
                pad_mask:         torch.Tensor    # (B, N) bool; True = ignore
               ) -> torch.Tensor:                 # (B, 1024)

        B = new_clip.shape[0]

        # Query: new image
        query_feat = self._project_image(new_clip, new_mobilenet)   # (B, D)
        query_seq  = query_feat.unsqueeze(1)                         # (B, 1, D)

        # Keys/Values: calibration tokens + prepended prior
        kv_tokens   = self._build_kv(calib_clips, calib_mobilenets, calib_scores)
        prior_valid = torch.zeros(B, 1, dtype=torch.bool, device=pad_mask.device)
        full_mask   = torch.cat([prior_valid, pad_mask], dim=1)      # (B, N+1)

        # Cross-attention stack
        x = query_seq
        for layer in self.cross_layers:
            q_normed  = layer["ln_q"](x)
            kv_normed = layer["ln_kv"](kv_tokens)
            attn_out, _ = layer["attn"](
                query            = q_normed,
                key              = kv_normed,
                value            = kv_normed,
                key_padding_mask = full_mask,
            )
            x = x + attn_out
            x = x + layer["ffn"](layer["ln_ffn"](x))

        context = x.squeeze(1)                                       # (B, D)

        # Predict BOLD
        combined  = torch.cat([query_feat, context], dim=-1)         # (B, 2D)
        bold_pred = self.predictor(combined)                          # (B, 1024)
        return bold_pred


def count_params(model: nn.Module) -> tuple:
    total     = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total, trainable


# ── CELL 6 : TRAINING UTILITIES ───────────────────────────────────────────────

class EarlyStopping:
    """Monitors training loss; stops after `patience` epochs without improvement."""
    def __init__(self, patience: int, min_delta: float = 1e-5):
        self.patience  = patience
        self.min_delta = min_delta
        self.best      = float("inf")
        self.counter   = 0

    def step(self, loss: float) -> bool:
        if loss < self.best - self.min_delta:
            self.best    = loss
            self.counter = 0
            return False
        self.counter += 1
        return self.counter >= self.patience

    def __repr__(self):
        return f"ES(best={self.best:.6f}, {self.counter}/{self.patience})"


class WarmupCosineLR(torch.optim.lr_scheduler._LRScheduler):
    """Linear warmup → cosine decay to min_lr_ratio × base_lr."""
    def __init__(self, optimizer, warmup_steps: int, total_steps: int,
                 min_lr_ratio: float = 0.05, last_epoch: int = -1):
        self.warmup_steps  = warmup_steps
        self.total_steps   = total_steps
        self.min_lr_ratio  = min_lr_ratio
        super().__init__(optimizer, last_epoch)

    def get_lr(self):
        s = self.last_epoch
        if s < self.warmup_steps:
            f = s / max(1, self.warmup_steps)
        else:
            progress = (s - self.warmup_steps) / max(1, self.total_steps - self.warmup_steps)
            f = self.min_lr_ratio + 0.5 * (1 - self.min_lr_ratio) * (
                1 + math.cos(math.pi * progress)
            )
        return [base * f for base in self.base_lrs]


# ── CELL 7 : TRAIN / EVAL LOOPS ───────────────────────────────────────────────

def train_one_epoch(model, loader, optimizer, scheduler, scaler, device) -> float:
    model.train()
    total_loss = 0.0
    n_batches  = 0

    for batch in tqdm(loader, desc="  train", leave=False):
        cc  = batch["calib_clips"].to(device,       non_blocking=True)
        cmn = batch["calib_mobilenets"].to(device,  non_blocking=True)
        cs  = batch["calib_scores"].to(device,      non_blocking=True)
        pm  = batch["pad_mask"].to(device,          non_blocking=True)
        tc  = batch["target_clips"].to(device,      non_blocking=True)
        tmn = batch["target_mobilenets"].to(device, non_blocking=True)
        tb  = batch["target_bolds"].to(device,      non_blocking=True)

        optimizer.zero_grad()

        if scaler is not None:
            with torch.cuda.amp.autocast():
                pred = model(tc, tmn, cc, cmn, cs, pm)
                loss = F.mse_loss(pred, tb)
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), CFG["grad_clip"])
            scaler.step(optimizer)
            scaler.update()
        else:
            pred = model(tc, tmn, cc, cmn, cs, pm)
            loss = F.mse_loss(pred, tb)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), CFG["grad_clip"])
            optimizer.step()

        scheduler.step()
        total_loss += loss.item()
        n_batches  += 1

    return total_loss / max(n_batches, 1)


@torch.no_grad()
def evaluate(model_for_eval: nn.Module,
             subject_data:   dict,
             device,
             n_calib:        int = 20,
             n_repetitions:  int = 10) -> dict:
    """
    Evaluate on each subject's held-out test images.

    Repeats n_repetitions times with fresh calibration sets and averages
    predictions for stability. Returns mean Pearson r across 1024 parcels
    per subject.
    """
    model_for_eval.eval()
    results = {}

    for subj, data in subject_data.items():
        test_mask  = data["is_test"]
        train_mask = ~test_mask

        if test_mask.sum() == 0 or train_mask.sum() < n_calib:
            continue

        test_clips      = data["clips"][test_mask]       # (T, 512)
        test_mobilenets = data["mobilenets"][test_mask]  # (T, 176)
        test_bolds      = data["bolds"][test_mask]       # (T, 1024)
        T               = len(test_clips)

        train_idx  = np.where(train_mask)[0]
        preds_reps = []

        for _ in range(n_repetitions):
            calib_idx = np.random.choice(train_idx, size=n_calib, replace=False)

            c_clips  = torch.FloatTensor(
                data["clips"][calib_idx]).unsqueeze(0).to(device)           # (1, n, 512)
            c_mns    = torch.FloatTensor(
                data["mobilenets"][calib_idx]).unsqueeze(0).to(device)      # (1, n, 176)
            c_scores = torch.FloatTensor(
                data["scores"][calib_idx]).unsqueeze(0).unsqueeze(-1).to(device)
            c_pad    = torch.zeros(1, n_calib, dtype=torch.bool).to(device)

            batch_preds = []
            for start in range(0, T, 64):
                end = min(start + 64, T)
                bs  = end - start
                tc  = torch.FloatTensor(test_clips[start:end]).to(device)
                tmn = torch.FloatTensor(test_mobilenets[start:end]).to(device)

                pred = model_for_eval(
                    tc,
                    tmn,
                    c_clips.expand(bs,  -1, -1),
                    c_mns.expand(bs,    -1, -1),
                    c_scores.expand(bs, -1, -1),
                    c_pad.expand(bs,    -1),
                )
                batch_preds.append(pred.cpu().numpy())

            preds_reps.append(np.vstack(batch_preds))

        preds_mean = np.mean(preds_reps, axis=0)   # (T, 1024)

        r_vals = []
        for p in range(1024):
            gt = test_bolds[:, p]
            if gt.std() > 0:
                r, _ = pearsonr(preds_mean[:, p], gt)
                r_vals.append(r)
        results[subj] = float(np.mean(r_vals)) if r_vals else 0.0

    return results


# ── CELL 8 : INITIALISE EVERYTHING ────────────────────────────────────────────

device = torch.device(CFG["device"])

print("\nBuilding datasets …")
train_ds = MultiViewMetaDataset(
    subject_data    = subject_data,
    n_calib_options = CFG["n_calib_options"],
    n_views         = CFG["n_views_per_image"],
    split           = "train",
)

train_loader = DataLoader(
    train_ds,
    batch_size         = CFG["batch_size"],
    shuffle            = True,
    num_workers        = CFG["num_workers"],
    collate_fn         = meta_collate,
    pin_memory         = True,
    drop_last          = True,
    persistent_workers = CFG["num_workers"] > 0,
)

steps_per_epoch = len(train_loader)
total_steps     = CFG["epochs"] * steps_per_epoch
warmup_steps    = CFG["warmup_epochs"] * steps_per_epoch

print(f"\nEpoch length : {steps_per_epoch} batches"
      f"  (effective batch: {CFG['batch_size'] * max(1, torch.cuda.device_count())})")
print(f"Total steps  : {total_steps}")
print(f"Warmup steps : {warmup_steps}")

# ── Model ─────────────────────────────────────────────────────────────────────
model = CalibratedBoldPredictor(CFG)
n_total, n_train_p = count_params(model)
print(f"\nModel parameters: {n_total:,}  (all trainable: {n_train_p:,})")
print(f"Predictor MLP   : {CFG['mlp_n_layers']} layers  "
      f"hidden={CFG['mlp_hidden']}")

n_gpus = torch.cuda.device_count()
if n_gpus > 1:
    print(f"DataParallel across {n_gpus} GPUs")
    model = nn.DataParallel(model)
model = model.to(device)

# ── Optimiser + Scheduler ─────────────────────────────────────────────────────
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr           = CFG["lr"],
    weight_decay = CFG["weight_decay"],
    betas        = (0.9, 0.98),
)

scheduler = WarmupCosineLR(
    optimizer,
    warmup_steps = warmup_steps,
    total_steps  = total_steps,
    min_lr_ratio = 0.05,
)

scaler = (torch.cuda.amp.GradScaler()
          if CFG["use_amp"] and device.type == "cuda"
          else None)

early_stop = EarlyStopping(
    patience  = CFG["early_stop_patience"],
    min_delta = CFG["early_stop_min_delta"],
)


# ── CELL 9 : TRAINING LOOP ────────────────────────────────────────────────────

history    = {"train_loss": [], "lr": []}
best_loss  = float("inf")
best_epoch = 0
best_state = None

print(f"\n{'═'*65}")
print(f"  Training  (up to {CFG['epochs']} epochs, "
      f"early stop patience={CFG['early_stop_patience']})")
print(f"{'═'*65}\n")

t_global = time.perf_counter()

for epoch in range(1, CFG["epochs"] + 1):
    t0 = time.perf_counter()

    train_loss = train_one_epoch(
        model, train_loader, optimizer, scheduler, scaler, device
    )

    lr_now  = optimizer.param_groups[0]["lr"]
    elapsed = time.perf_counter() - t0

    history["train_loss"].append(train_loss)
    history["lr"].append(lr_now)

    improved = "✓" if train_loss < best_loss else " "
    print(f"Epoch {epoch:3d}/{CFG['epochs']}  "
          f"loss={train_loss:.6f} {improved}  "
          f"lr={lr_now:.2e}  "
          f"({elapsed:.0f}s)  "
          f"{early_stop}")

    if train_loss < best_loss:
        best_loss  = train_loss
        best_epoch = epoch
        actual     = model.module if isinstance(model, nn.DataParallel) else model
        best_state = {k: v.cpu().clone() for k, v in actual.state_dict().items()}
        torch.save({
            "epoch":       epoch,
            "train_loss":  best_loss,
            "model_state": best_state,
            "cfg":         CFG,
        }, os.path.join(CFG["output_dir"], "best_model.pt"))

    if early_stop.step(train_loss):
        print(f"\n⚡ Early stopping at epoch {epoch}. "
              f"Best epoch: {best_epoch}  loss: {best_loss:.6f}")
        break

total_min = (time.perf_counter() - t_global) / 60
print(f"\n✓ Training done in {total_min:.1f} min")
print(f"  Best epoch: {best_epoch}   best loss: {best_loss:.6f}")


# ── CELL 10 : FINAL EVALUATION ────────────────────────────────────────────────

print("\nLoading best checkpoint …")
ckpt = torch.load(
    os.path.join(CFG["output_dir"], "best_model.pt"),
    map_location=device,
)

eval_model = CalibratedBoldPredictor(CFG).to(device)
eval_model.load_state_dict(ckpt["model_state"])

print(f"Evaluating on held-out test images …  "
      f"(n_calib={CFG['n_calib_eval']}, n_repetitions=10)")

pearson_results = evaluate(
    eval_model,
    subject_data,
    device,
    n_calib       = CFG["n_calib_eval"],
    n_repetitions = 10,
)

print("\n── Per-subject mean Pearson r (all 1024 DiFuMo parcels) ─────────────")
all_r = []
for subj, r in sorted(pearson_results.items()):
    print(f"  {subj}: {r:.4f}")
    all_r.append(r)
mean_r = float(np.mean(all_r)) if all_r else 0.0
print(f"  ───────────────")
print(f"  Mean : {mean_r:.4f}")

history["pearson_per_subject"] = pearson_results
history["mean_pearson"]        = mean_r
history["best_epoch"]          = best_epoch
history["best_train_loss"]     = best_loss

with open(os.path.join(CFG["output_dir"], "training_history.json"), "w") as f:
    json.dump(history, f, indent=2)


# ── CELL 11 : PLOTS ───────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 3, figsize=(18, 4))

ax = axes[0]
epochs_x = range(1, len(history["train_loss"]) + 1)
ax.plot(epochs_x, history["train_loss"], "#2563EB", lw=1.5)
ax.axvline(best_epoch, color="#DC2626", ls="--", lw=1.2,
           label=f"Best epoch {best_epoch}")
ax.set_xlabel("Epoch");  ax.set_ylabel("MSE (z-scored BOLD)")
ax.set_title("Training Loss");  ax.legend();  ax.grid(alpha=0.3)

ax = axes[1]
ax.plot(epochs_x, history["lr"], "#16A34A", lw=1.5)
ax.set_xlabel("Epoch");  ax.set_ylabel("Learning Rate")
ax.set_title("LR Schedule (warmup + cosine)");  ax.grid(alpha=0.3)

ax     = axes[2]
subjs  = sorted(pearson_results)
r_vals = [pearson_results[s] for s in subjs]
colors = ["#2563EB", "#16A34A", "#D97706", "#9333EA"]
bars   = ax.bar(subjs, r_vals, color=colors[:len(subjs)], zorder=3)
ax.axhline(mean_r, color="#DC2626", ls="--", lw=1.5,
           label=f"Mean = {mean_r:.3f}")
for bar, val in zip(bars, r_vals):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.003,
            f"{val:.3f}", ha="center", va="bottom", fontsize=10)
ax.set_ylabel("Mean Pearson r (1024 parcels)")
ax.set_title("BOLD Prediction Accuracy\n(30 held-out test images)")
ax.legend();  ax.grid(alpha=0.3, axis="y", zorder=0)

plt.tight_layout()
plot_path = os.path.join(CFG["output_dir"], "training_summary.png")
plt.savefig(plot_path, dpi=150, bbox_inches="tight")
plt.show()
print(f"Plot saved → {plot_path}")


# ── CELL 12 : YEO-7 BREAKDOWN ─────────────────────────────────────────────────

YEO_NAMES = {
    1: "Visual", 2: "Somatomotor", 3: "DorsAttn",
    4: "VentAttn", 5: "Limbic", 6: "FPN", 7: "DMN",
}

print("\n── Yeo-7 network breakdown (mean Pearson r per network) ─────────────")
print(f"  {'Network':<14}", end="")
for subj in sorted(pearson_results):
    print(f"  {subj:>7}", end="")
print()

eval_model.eval()

for net_id, net_name in sorted(YEO_NAMES.items()):
    net_mask = (YEO_ASSIGN == net_id)
    n_p      = int(net_mask.sum())
    print(f"  {net_name:<14}(n={n_p:3d}) ", end="")

    for subj in sorted(subject_data):
        data      = subject_data[subj]
        test_mask = data["is_test"]
        if test_mask.sum() == 0:
            print(f"  {'n/a':>7}", end="")
            continue

        test_clips_t    = data["clips"][test_mask]
        test_mns_t      = data["mobilenets"][test_mask]
        test_bolds_net  = data["bolds"][test_mask][:, net_mask]
        train_idx       = np.where(~test_mask)[0]

        calib_n   = min(CFG["n_calib_eval"], len(train_idx))
        calib_idx = np.random.choice(train_idx, size=calib_n, replace=False)

        c_clips  = torch.FloatTensor(
            data["clips"][calib_idx]).unsqueeze(0).to(device)
        c_mns    = torch.FloatTensor(
            data["mobilenets"][calib_idx]).unsqueeze(0).to(device)
        c_scores = torch.FloatTensor(
            data["scores"][calib_idx]).unsqueeze(0).unsqueeze(-1).to(device)
        c_pad    = torch.zeros(1, calib_n, dtype=torch.bool).to(device)

        T     = int(test_mask.sum())
        preds = []
        with torch.no_grad():
            for start in range(0, T, 64):
                end = min(start + 64, T)
                bs  = end - start
                tc  = torch.FloatTensor(test_clips_t[start:end]).to(device)
                tmn = torch.FloatTensor(test_mns_t[start:end]).to(device)
                pred = eval_model(
                    tc, tmn,
                    c_clips.expand(bs,  -1, -1),
                    c_mns.expand(bs,    -1, -1),
                    c_scores.expand(bs, -1, -1),
                    c_pad.expand(bs,    -1),
                )
                preds.append(pred.cpu().numpy())

        preds_np   = np.vstack(preds)[:, net_mask]
        r_vals_net = [
            pearsonr(preds_np[:, p], test_bolds_net[:, p])[0]
            for p in range(n_p)
            if test_bolds_net[:, p].std() > 0
        ]
        net_r = float(np.mean(r_vals_net)) if r_vals_net else 0.0
        print(f"  {net_r:>7.4f}", end="")
    print()

print("\n✓ All done.")
print(f"\nSaved artefacts in {CFG['output_dir']}:")
for p in sorted(Path(CFG["output_dir"]).iterdir()):
    print(f"  {p.name:<35} {p.stat().st_size / 1e6:.2f} MB")
```
